А яка найпопулярніша зарплата серед інженерів?

```python
# scipy
from scipy import stats
most_popular_salary = stats.mode(salaries, keepdims=False)
print(f"Scipy mode result:\n {most_popular_salary}")  # always 1 value, dunno how ot bypass in scipy

# numpy
unique_counts = np.unique(salaries, return_counts=True)
mode_count = np.amax(unique_counts[1])
stacked = np.dstack(unique_counts)[0]
modes = stacked[np.in1d(stacked[:, 1], np.asarray([mode_count]))]
print(f"Numpy mode result:\n {modes[:, 0]}")
```

<font color='blue'>**[Optional]**</font> Щоб відпрацювати знання лінійнох алгебри напишіть функцію на чистому Python, яка би повертала матричний добуток двох матриць. Також, зверніть увагу наскільки легше це зробити за допомогою Numpy

```python
def matr_product(x, y):
  # Expected rows of one matrix have the same length, no broadcasting
  x_rows = len(x)
  x_cols = len(x[0])
  y_rows = len(y)
  y_cols = len(y[0])

  assert x_cols == y_rows,  f"Insuitable shapes for multiplication"

  result = [[0 for i in range(y_cols)] for j in range(x_rows)]

  for i in range(x_rows):
    for j in range(y_cols):
      for k in range(x_cols):
        result[i][j] += x[i][k]*y[k][j]

  return result
```

```python
#use this code to test your solution 
a = np.array([
    [1, 2],
    [3, 4]
])
b = np.array([
    [11, 12], 
    [13, 14]
])
assert matr_product(a, b)[0][0] == 37,  f"You calculation is wrong"
```
